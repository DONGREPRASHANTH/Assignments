### task 3 Clustering with Original Data:

```python
import warnings

# Suppress all warnings
warnings.filterwarnings("ignore")
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"
```

#### Applying a clustering algorithm (K-means) to the original dataset.

```python
from sklearn.cluster import KMeans

# Assuming your dataset is in a variable called X (shape: samples x features)
k = 3  # Choose number of clusters; can be tuned or selected based on methods like elbow or silhouette
Wine=pd.read_csv('wine.csv')
kmeans = KMeans(n_clusters=k, random_state=42)
cluster_labels = kmeans.fit_predict(Wine)
```

#### Visualizing the clustering results using appropriate plots.

```python
# Ploting the clusters
# Visualize using "Alcohol" and "Malic" features
plt.figure(figsize=(10, 6))
sns.scatterplot(x=Wine['Alcohol'], y=Wine['Malic'], hue=cluster_labels, palette='Set1', s=100)
plt.title('K-means Clustering Results')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Visualize using "Alcohol" and "Malic" features
plt.figure(figsize=(10, 6))
sns.scatterplot(x=Wine['Flavanoids'], y=Wine['Nonflavanoids'], hue=cluster_labels, palette='Set1', s=100)
plt.title('K-means Clustering Results')
plt.xlabel('Flavanoids')
plt.ylabel('Nonflavanoids')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

#### Evaluating the Clustering Performance
we are using the silhouette score and the Davies–Bouldin index to evaluate the clustering performance.

#### Silhouette Score
The silhouette score measures how similar an object is to its own cluster compared to other clusters. The score ranges from -1 to 1, where a higher score indicates better-defined clusters.

```python
from sklearn.metrics import silhouette_score

# Calculating the silhouette score
silhouette_avg = silhouette_score(Wine, cluster_labels)
print(f'Silhouette Score: {silhouette_avg:.2f}')
```

#### Davies–Bouldin Index
The Davies–Bouldin index is another metric for evaluating clustering performance, where a lower score indicates better clustering.

```python
from sklearn.metrics import davies_bouldin_score

# Calculate the Davies–Bouldin index
db_index = davies_bouldin_score(Wine, cluster_labels)
print(f'Davies–Bouldin Index: {db_index:.2f}')
```

3. Reflect on the Impact of Dimensionality Reduction on Clustering Performance we have

Positive Impact:

Noise Reduction: PCA helps in reducing noise by focusing on the components that explain the most variance, which can lead to clearer clusters.
Improved Visualization: Reducing dimensions to 2 allows for a more straightforward interpretation of clustering resuts.
Negative Impact:

Information Loss: Some important information may be lost during PCA, which could affect clustering outcomes if critical features are discarded.
Cluster Overlap: In some cases, PCA may cause clusters that were distinct in high-dimensional space to overlap, leading to potential misclassification.

4. Analyzing the Trade-offs Between Using PCA and Clustering Directly on the Original Dataset

PCA Advantages:

Computational Efficiency: Reduces the number of features, leading to faster computations during clustering.
Focus on Variance: PCA emphasizes the most significant features, which can enhance clustering performance by eliminating less informative features.

PCA Disadvantages:

Interpretability: The principal components are linear combinations of the original features, making it harder to interpret the results in terms of original features and Loss of Detail

Original Dataset Advantages:

Direct Interpretation: Clustering results can be directly related to original features, which can be more meaningful for domain-specific analysis 
and Full Information Utilization
Original Dataset Disadvantages:

Computational Complexity: Higher dimensionality can lead to longer computation times and may require more sophisticated algorithms to handle potential overfitting.
